```python
# top=75
# threshold=2

import pandas as pd
deals = pd.read_csv("dealsdata.csv", encoding='latin-1')
MandA = deals[deals['Share Size']=='100%']
ma2005 = MandA[MandA['Year']==2005]
ma2010 = MandA[MandA['Year']==2010]
ma2019 = MandA[MandA['Year']==2018]
```

Can show that M&A network resembles that of BRI (govt preference indicators) and WFOE (company preference indicators)

```python
deals.describe()
```

```python
# all years, Deals general

colourTony = 'black'
colourTony = 'white'

import networkx as nx 
import matplotlib.pyplot as plt
import itertools

from matplotlib.pyplot import figure
figure(num=None, figsize=(25, 25), dpi=120, facecolor='w', edgecolor='k')

G = nx.Graph() 

for index, row in ma2005.iterrows():
    G.add_edge("China",row[deals.columns[-4]], weight=row[deals.columns[3]])
#     G.add_edge("Hong Kong, China",row[joined_df.columns[0]], weight=row[joined_df.columns[10]], color=colourTony)
#     if row[joined_df.columns[10]]-row[joined_df.columns[9]]>0:
#         if row[joined_df.columns[-1]]-row[joined_df.columns[-2]]<=0:
#             counter=0
#             for i in list(range(1,10+1)):
#                 if row[joined_df.columns[i]]-row[joined_df.columns[i-12+1]]>0:
#                     counter+=1
#             if counter>=threshold:
#                 G.add_edge("Hong Kong, China",row[joined_df.columns[0]], weight=row[joined_df.columns[10]], color='green')


color_map = []
for node in G:
    if node == "China":
        color_map.append('green')
    
    else:
        color_map.append('yellow')

# pos = nx.spring_layout(G, k=0.15,iterations=50)
pos = nx.circular_layout(G)

edges = G.edges()
# colors = [G[u][v]['color'] for u,v in edges]
weights = [G[u][v]['weight'] for u,v in edges]

# nx.draw(G, pos=pos, with_labels=True, node_size = 1000, font_size=30, edges=edges, edge_color=colors, node_color = color_map)
nx.draw(G, pos=pos, with_labels=True, node_size = 1000, font_size=30, edges=edges, node_color = color_map)
```
